```python
import json
import os

BASE_DIR = os.path.dirname(os.path.abspath("__file__"))
THEO_PATH = os.path.join(BASE_DIR, "questions_theoriques.json")
PROB_PATH = os.path.join(BASE_DIR, "problemes.json")
```

```python
def charger(path):
    with open(path, "r", encoding="utf-8") as f:
        return json.load(f)

def sauvegarder(data, path):
    with open(path, "w", encoding="utf-8") as f:
        json.dump(data, f, ensure_ascii=False, indent=2)

def ajouter_questions(path, chapitre, questions):
    """Ajoute des questions à un fichier JSON.
    
    questions: liste de dicts {"id": "X.Y", "question": "...", "reponse": "..."}
    """
    data = charger(path)
    for q in questions:
        q["chapitre"] = chapitre
        data.append(q)
    sauvegarder(data, path)
    print(f"{len(questions)} question(s) ajoutée(s) au chapitre '{chapitre}'.")

def lister_chapitres(path):
    data = charger(path)
    chapitres = {}
    for q in data:
        ch = q["chapitre"]
        chapitres[ch] = chapitres.get(ch, 0) + 1
    for ch, n in chapitres.items():
        print(f"  {ch}: {n} questions")
    return chapitres
```

```python
print("=== Questions théoriques ===")
lister_chapitres(THEO_PATH)
print("\n=== Problèmes ===")
lister_chapitres(PROB_PATH)
```
